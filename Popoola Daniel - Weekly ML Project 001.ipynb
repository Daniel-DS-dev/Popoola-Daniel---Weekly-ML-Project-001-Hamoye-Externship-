# PART 1: Load Data

```python
#Importing the data
cell_phones = pd.read_csv("https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv")
population = pd.read_csv("https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv")
country_code = pd.read_csv("https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/b557bdc0693112f45c2d0330bb42a54b37713ac9/ddf--entities--geo--country.csv")
```

# PART 2: Join Data
* Join the ‘Cell phones’ and ‘population’ dataframe (With an inner join on ‘Country’ and ‘year’)

```python
phone_population = pd.merge(cell_phones, population, how = 'inner') 
```

```python
phone_population.head()
```

* In the ‘Geo country codes’ rename the columns{‘country to ‘geo’, and ‘name’ to ‘Country’

```python
country_code = country_code.rename(columns={'country':'geo', 'name':'Country'})
```

```python
country_code.head()
```

* Then, select the ‘geo’ and ‘Country’ columns from the ‘geo_country’ dataframe, and join with your ‘merged population and cell phone’ data.

```python
geo_phn_pop = pd.merge(phone_population, country_code, how = 'inner')
```

```python
geo_phn_pop.head()
```

```python
col = phone_population.columns

cols = [] 
for i in col:
    cols.append(i)
    
cols.append('geo')
cols
```

```python
#Selecting only the columns we need
data = geo_phn_pop[cols]
```

```python
data.head()
```

* In 2017, what were the top 5 countries with the most cell phones total?
* We will get this by aggregating through our data.

```python
#filtering our data to show only data in the year 2017.
data_2017 = data[data['year'] == 2017]
data_2017.head()
```

```python
#Grouping each country together, and getting the total cell phones per country in 2017
data_2017_grouped = data_2017[['Country', 'cell_phones_total']].groupby('Country').sum()
data_2017_grouped.head()
```

```python
#Now sorting this data from highest to lowest, and getting the top 5 countries with the highest number of cell phones in 2017
data_2017_grouped_sorted = data_2017_grouped.sort_values(by='cell_phones_total', ascending = False)

data_2017_grouped_sorted.head(5)
```

* The countries with the highest number of cell phones in 2017 are China, India, Indonesia, United States, and Russia, in that order.
